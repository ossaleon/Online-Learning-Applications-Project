## Plots optimality

```python
# PLOTS USERCLASS 1
temp_user = C1

opt_idx_price = np.argmax(temp_user.conversion_rate_function(PRICES) * (PRICES - MARGIN_PARAM)) #optimal value of conversion rate between the arms
opt_price = PRICES[opt_idx_price]

opt_conversion_rate = temp_user.conversion_rate_function(opt_price)

opt_bid_arm = np.argmax(temp_user.n_daily_clicks_function(BIDS) * opt_conversion_rate * (PRICES[opt_idx_price] - MARGIN_PARAM) - temp_user.cum_daily_costs_function(BIDS)) #optimal value of conversion rate between the arms
opt_bid = BIDS[opt_bid_arm]

opt_n_daily_clicks = temp_user.n_daily_clicks_function(opt_bid)
opt_cum_daily_costs = temp_user.cum_daily_costs_function(opt_bid)


opt = opt_n_daily_clicks * opt_conversion_rate * (PRICES[opt_idx_price] - MARGIN_PARAM) - opt_cum_daily_costs




#Plot

plt.figure(figsize=(8,3))
plt.suptitle('user class C1')
for price in PRICES:
    plt.plot(BIDS, temp_user.n_daily_clicks_function(BIDS) * temp_user.conversion_rate_function(price) * (price - MARGIN_PARAM) - temp_user.cum_daily_costs_function(BIDS), label = f'price={price}')
plt.plot(opt_bid, opt, 'o', color = "C9", label = "opt")
plt.xlabel('bid')
plt.ylabel('reward')
plt.legend(fontsize=8)
plt.grid(True)
```
